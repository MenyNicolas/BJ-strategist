```python
import numpy as np
import pandas as pd
import joblib
```

```python
df = pd.read_csv(r"../data/strategy_chart/basic_strategy_splitting.csv")
```

```python
df_validation = df.copy()
```

```python
for index, row in df.iterrows():  # ✅ df est ton DataFrame complet
    action = None
    df.at[index, "action"] = action
```

```python
def model_predict(total_joueur, carte_dealer):
    model = joblib.load("../model/splitting_model.pkl")
    label_encoder = joblib.load("../model/splitting_label_encoder.pkl")

    X = np.array([[total_joueur, carte_dealer]])  # ✅ 2D input

    pred_encoded = model.predict(X)[0]
    action = label_encoder.inverse_transform([pred_encoded])[0]

    return action
```

```python
for index, row in df.iterrows():  # ✅ df est ton DataFrame complet
    action = model_predict(row.player_total, row.dealer_card)
    df.at[index, "action"] = action
```

```python
# Vérifier si les colonnes sont identiques
comparaison = (df['action'] == df_validation['action']).all()
print(f"Les colonnes sont-elles identiques ? {comparaison}")

# Afficher les différences si elles existent
if not comparaison:
    print("\nDifférences trouvées :")
    diff_indices = df[df['action'] != df_validation['action']].index
    for idx in diff_indices:
        print(f"Index {idx}: df={df.at[idx, 'action']}, df_validation={df_validation.at[idx, 'action']}")
```
