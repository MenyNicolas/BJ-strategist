```python
import numpy as np
import pandas as pd
import joblib
```

```python
df = pd.read_csv(r"../data/strategy_chart/basic_strategy.csv")
```

```python
df_validation = df.copy()
```

```python
for index, row in df.iterrows():  # ✅ df est ton DataFrame complet
    action = None
    df.at[index, "action"] = action
```

```python
def model_stand_hit_double(total_joueur, carte_dealer, is_soft):

    model = joblib.load("../model/basic_strategy_model.pkl")
    label_encoder = joblib.load("../model/basic_strategy_label_encoder.pkl")

    X = np.array([[total_joueur, carte_dealer, is_soft]])

    pred_encoded = model.predict(X)[0]
    action = label_encoder.inverse_transform([pred_encoded])[0]

    return action
```

```python
for index, row in df.iterrows():  # ✅ df est ton DataFrame complet
    action = model_stand_hit_double(row.player_total, row.dealer_card, row.is_soft)
    df.at[index, "action"] = action
```
